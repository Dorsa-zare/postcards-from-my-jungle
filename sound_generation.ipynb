```python
!pip install diffusers transformers accelerate scipy

import torch
from diffusers import AudioLDMPipeline
import scipy.io.wavfile as wav
import os

device = "cuda" if torch.cuda.is_available() else "cpu"

# Load pretrained AudioLDM model
pipe = AudioLDMPipeline.from_pretrained("cvssp/audioldm")
pipe = pipe.to(device)

# Create output folder
os.makedirs("sounds", exist_ok=True)

# Prompts for 4 jungle creature sounds
prompts = [
    "Deep jungle creature roar with echo, 6 seconds",
    "Wet amphibian clicking in humid rainforest, 4 seconds",
    "Distorted bird-like jungle call with reverb, 8 seconds",
    "Soft glowing bioluminescent creature fluttering in forest ambience, 5 seconds"
]

# Generate and save sounds
for i, prompt in enumerate(prompts):
    audio = pipe(prompt, num_inference_steps=50, audio_length_in_s=6).audios[0]
    wav.write(f"sounds/sound{i+1}.wav", rate=16000, data=audio)
    print(f"Saved sound{i+1}.wav")

print("All sounds generated!")
```
